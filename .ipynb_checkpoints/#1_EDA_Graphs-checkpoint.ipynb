## Graph plot details
1. Scatterplot - with & without Hue
2. Boxplot
3. Countplot
4. Pairplot
5. Heatmap
6. 3d matplot of Scatterplot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
import os
```

```python
os.getcwd()
```

```python
df = pd.read_csv("./Data/hearing_test.csv")
df.head(5)
```

# Scatter plot with & without Hue

```python
fig, axs = plt.pyplot.subplots(nrows=1, ncols=2,figsize=(7,3))
sns.scatterplot(data=df,x='age',y='physical_score',ax=axs[0])
sns.scatterplot(data=df,x='age',y='physical_score',hue='test_result',ax=axs[1])
```

# Regression plot

```python
plt.pyplot.figure(figsize=(7,3))
sns.regplot(data=df,x='age',y='physical_score',
            line_kws={"color":'r',"lw":1,"ls":'--','alpha':1}, #Alpha: Transperancy [0-1]
            scatter_kws={"color":'green',"marker":'x',"alpha":0.5,'s':10} #s: Marker size
           )
```

# Boxplot
'''
Distribution plot Min,Q1,Median,Q3,Max
Min & Max are above & below outliers
Outliers are "< Q1 - 1.5*IQR" and "> Q3 + 1.5*IQR"
IQR = Q3-Q1
'''

```python
fig,axs = plt.pyplot.subplots(nrows=2,ncols=2,figsize=(10,5))
sns.boxplot(data=df,ax=axs[0,0])
sns.boxplot(data=df,x='age',ax=axs[0,1])
sns.boxplot(data=df,x='age',y='physical_score',ax=axs[0,1])
sns.boxplot(data=df,x='test_result',y='age',ax=axs[1,0])
sns.boxplot(data=df,x='test_result',y='physical_score',ax=axs[1,1])
```

# Count plot

```python
plt.pyplot.figure(figsize=(10,5))
plt.pyplot.xticks(rotation=90)
sns.countplot(data=df,x='age')
```
